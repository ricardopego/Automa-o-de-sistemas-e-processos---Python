# Automação de Sistemas e Processos com Python

### Desafio:

Para controle de custos, todos os dias, seu chefe pede um relatório com todas as compras de mercadorias da empresa.
O seu trabalho, como analista, é enviar um e-mail para ele, assim que começar a trabalhar, com o total gasto, a quantidade de produtos compradas e o preço médio dos produtos.

E-mail do seu chefe: para o nosso exercício, coloque um e-mail seu como sendo o e-mail do seu chefe<br>
Link de acesso ao sistema da empresa: https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
import pyautogui
import time

# pyautogui.click -> clique com o mause
# pyautogui.write -> escrever um texto - obs: não escreve caracter especial ó,á,Â
# pyautogui.press -> apertar uma tecla
# pyautogui.hotkey -> apertar uma combinação de teclas
#time.sleep() -> pausa entre uma execuçãoe outra
# pyautogui.PAUSE = 1 -> espera o tempo solicitado entre todas as tarefas sem precisar do time sleep
# Passo a passo 

# Passo 1: entrar no sistema da empresa (link);
pyautogui.alert('Por favor não clique no e não mexa no mouse !')
pyautogui.PAUSE = 1
#pyautogui.hotkey('win','d')
#pyautogui.press ('win')
#pyautogui.write ('chorme')
#pyautogui.press ('enter')
pyautogui.hotkey ('ctrl','t')
pyautogui.write  ('https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema ')
pyautogui.press  ('enter')

#pode ser que o navegador precise carregar
time.sleep(1)

# Passo 2: fazer login;
#clicar no campo de login:

pyautogui.click(x=812, y=478)
pyautogui.write('meu_login')

pyautogui.click(x=710, y=571)
pyautogui.write('minha_senha')

pyautogui.click(x=763, y=657)
time.sleep(3)


# Passo 3: exportar a base de dados;
pyautogui.click(x=435, y=426) # clica no arquivo
pyautogui.click(x=877, y=259) # clica nos 3 pontos 
pyautogui.click(x=1019, y=741) #faz o dowload
time.sleep(2)

```

```python
# Passo 4: calcular os indicadores;

import pandas as pd

tabela = pd.read_csv(r'C:\Users\ricar_5z1c68x\Downloads\Compras.csv' , sep = ';')
display(tabela)

total_gasto = tabela ['ValorFinal'] . sum()
quantidade = tabela ['Quantidade'] . sum()
valor_medio = total_gasto / quantidade

print (total_gasto)
print (quantidade)
print (valor_medio)
```

```python
# Passo 5: enviar um e-mail para meu chefe;

# entrar no email

import pyperclip # seleciona um texto e copia - usado para texto com caracter especial

pyautogui.hotkey('ctrl','t')
pyautogui.write('https://mail.google.com/mail/u/0/?tab=rm&ogbl#inbox')
pyautogui.press('enter')
time.sleep(3)
                
# clicar no botao escrever 
pyautogui.click(x=146, y=248)
time.sleep(2)
                
# preencher as informaçoes do email
 #obs: ,.2f é para deixar duas casas decimais depois da virgula               
pyautogui.write('julyana.pego@gmail.com') #destinatario
pyautogui.press('enter')
time.sleep(1)
pyautogui.press('tab')                

pyperclip.copy('Relatório de Vendas') #assunto
pyautogui.hotkey('ctrl','v')
time.sleep(1)
pyautogui.press('tab')

texto = f'''
Prezado chefe

Segue o relatório o resultados dos seguintes relatórios

Total gasto: R$ {total_gasto:,.2f} 

Quantidade total de produtos: {quantidade:,.2f}

Preço médio: R$ {valor_medio:,.2f}

Qualquer dúvida é só chamar !

atenciosamente, Ricardo Pego Pereira.


'''
pyperclip.copy(texto) # corpo do e-mail
pyautogui.hotkey('ctrl','v')
time.sleep(1)             
# enviar
                
pyautogui.hotkey('ctrl','enter')
time.sleep(3)

pyautogui.hotkey('win','d')
time.sleep(1)

pyautogui.alert('E-mail enviado com sucesso !')
```
